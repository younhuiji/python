```python
# 일본의 연도별 기대 수명
df[df.country == 'Japan'][['country', 'year', 'lifeExp']]
```

```python
# 르완다의 연도별 기대수명
df[df.country == 'Rwanda'][['country', 'year', 'lifeExp']]
```

```python
for y in df.year.unique():
    print(y, ':' ,df[df.year == y]['lifeExp'].mean())
```

```python
# 연도별 기대수명 평균
df.groupby('year').lifeExp.mean()
```

```python
for y in df.year.unique():
    print(y, ':' ,df[df.year == y]['gdpPercap'].mean())
```
